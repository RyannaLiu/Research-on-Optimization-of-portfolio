```python
import numpy as np
import pandas as pd
import scipy.stats as scs
```

```python
data=pd.read_excel('hw1.xlsx')
```

```python
data=pd.DataFrame(data)
```

```python
data.set_index(['Date'],inplace=True)
```

```python
data.index
```

```python
data_pct=data.pct_change()
```

```python
data_pct=data_pct.iloc[1:,:]
data_pct
```

```python
pct_mean=pd.DataFrame(data_pct.mean(),columns=['mean']).T
pct_std=pd.DataFrame(data_pct.std(),columns=['std']).T
pct_skew=pd.DataFrame(data_pct.skew(),columns=['skew']).T
```

```python
sta=pd.concat([pct_mean,pct_std,pct_skew])
sta
```

```python
SPTR=data_pct.iloc[:,1]
```

```python
scs.jarque_bera(SPTR)
```

```python
def w_jb_test(x):
    return scs.jarque_bera(x)[0]
def p_jb_test(x):
    return scs.jarque_bera(x)[1]
```

```python
w=pd.DataFrame(data_pct.apply(w_jb_test),columns=['w']).T
p=pd.DataFrame(data_pct.apply(p_jb_test),columns=['p']).T
```

```python
w
```

```python
results=pd.concat([pct_mean,pct_std,pct_skew,w,p])
```
