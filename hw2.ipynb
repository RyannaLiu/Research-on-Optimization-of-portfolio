```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import lag_plot
```

```python
data_pct=pd.read_excel('data_pct.xlsx')
```

```python
data_pct
data_pct.set_index(['Date'], inplace=True)
```

```python
data_pct
```

```python
def f(x):
    
    fig,axes=plt.subplots(1,3,figsize=(15,5))
    plt.title([x.index])

    
    ax=plt.subplot(131)
    x.iloc[0:2258].hist()


    plt.subplot(132)
    x.iloc[2258:].hist()

    plt.subplot(133)
    lag_plot(x)

```

```python
data_pct.apply(f)
```
